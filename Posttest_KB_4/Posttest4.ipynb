## Data Cleaning

```python
from sklearn.impute import SimpleImputer
```

```python
df.isna().sum()
```

## Ubah nilai null yang memiliki tipe data kategorikal

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["Potential"] = imputer.fit_transform(df[["Potential"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

## Ubah nilai null yang memiliki tipe data desimal

```python
imputer = SimpleImputer(strategy = 'mean')
df["GK Rating"] = imputer.fit_transform(df[["GK Rating"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

## Ubah nilai null yang memiliki tipe data bilang bulat

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["RB Rating"] = imputer.fit_transform(df[["RB Rating"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

## Ubah tipe data angka menjadi kategorikal

```python
from sklearn.preprocessing import OrdinalEncoder 
```

```python
encoder = OrdinalEncoder() 
df["LB Rating"] = encoder.fit_transform(df[["LB Rating"]]) 
df.head()
```

## One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
encoder = OneHotEncoder(sparse=False) 
PotentialEnc = encoder.fit_transform(df[["Potential"]]) 
PotentialEnc = pd.DataFrame(PotentialEnc) 
df = df.join(PotentialEnc) 
df 
```
